```python
import numpy as np
import os 
import PIL
import PIL.Image 
import tensorflow as tf
# import tensorflow_datasets as tfds
import pathlib
```

```python
dataset_url = "https://github.com/JakeMWu/cv_num_cards_deck/blob/main/14-39/Images.tar.gz?raw=True"
archive_path = tf.keras.utils.get_file(origin=dataset_url, fname="Images.tar.gz", extract=True)

# The directory where the files are extracted
# Note: tf.keras.utils.get_file extracts to a cache subdirectory, so we need to adjust the path.
extracted_folder_path = pathlib.Path(archive_path).parent / "Images" # Adjust the "Images" folder name as necessary based on the actual structure

# Check if the path adjustment is needed
if not extracted_folder_path.exists():
    # Fall back to the parent directory if the expected one doesn't exist
    extracted_folder_path = pathlib.Path(archive_path).parent

# List all image files
# Adjust the glob pattern as necessary based on the file types and structure
image_files = list(extracted_folder_path.glob('**/*.jpg')) # Example pattern, adjust as needed

image_count = len(image_files)
print(f"Total images: {image_count}")
```
